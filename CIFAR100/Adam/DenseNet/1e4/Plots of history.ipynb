```python
import pickle
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
def plot(history, sharey=False):
    n = len(history.history['acc']) + 1
    fig, ax = plt.subplots(ncols=2, figsize=(10, 4), sharey=sharey)

    ax[0].set_ylabel('Accuracy');
    ax[0].set_xlabel('Epochs');
    lines = ax[0].plot(range(1, n, 5), history.history['acc'][::5], 'r-',
                       range(1, n, 5), history.history['val_acc'][::5], 'g-');
    ax[0].legend(lines, ('Training', 'Validation'));

    ax[1].set_xlabel('Epochs')
    ax[1].set_ylabel('Loss')
    lines = ax[1].plot(range(1, n, 5), history.history['loss'][::5], 'r-',
                       range(1, n, 5), history.history['val_loss'][::5], 'g-');
    ax[1].legend(lines, ('Training', 'Validation'))

    fig.tight_layout()
```
